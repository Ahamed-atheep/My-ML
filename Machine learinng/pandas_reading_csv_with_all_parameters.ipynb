```python
from io import StringIO,BytesIO
```

```python
data=('col1,col2,col3\n'
     'x,y,1\n'
     'a,b,2\n'
     'c,d,3')
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))
```

```python
## Read from Spefic Columns
df=pd.read_csv(StringIO(data),usecols=['col1','col3'])
df.to_csv('file.csv')
```

```python
## Specifing Columns data types

data=('a,b,c,d\n'
      '1,2,3,4\n'
      '5,6,7,8\n'
      '9,10,11,12')
```

```python
print(data)
```

```python
df=pd.read_csv(StringIO(data),dtype=object) ## Here object is string
```

```python
df['a']
```

```python
## spefy the different datatype in different cols

df=pd.read_csv(StringIO(data),dtype={'b':int,'c':float,'a':'Int64'})
```

```python
df.head()
```

```python
df['a'][1]
```

```python
type(df['a'][1])
```

```python
data=('index,a,b,c \n'
      '4,apple,bat,5.7\n'
      '8,orange,cow,10')
```

```python
pd.read_csv(StringIO(data),index_col=0)
```

```python
data=('a,b,c\n'
      '4,apple,but\n'
      '8,orange,cow'
     )
```

```python
pd.read_csv(StringIO(data),index_col=False)
```

```python
## combining use of Usecols and index_col
data=('a,b,c\n'
     '4,apple,but\n'
     '8,orange,cow')
```

```python
pd.read_csv(StringIO(data),usecols=['b','c'],index_col=False)
```

```python
## quoting and Escape Characters use in NLP

data='a,b\n "Hello, \\"Bob\\",nice to see you",5'
```

```python
pd.read_csv(StringIO(data),escapechar='\\')
```
