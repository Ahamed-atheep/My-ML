```python
import matplotlib.pyplot as plt
%matplotlib inline ## don't need to write the plt.show()
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
a=np.arange(40,50)
b=np.arange(50,60)
len(x)==len(y)
```

```python
## Scatter plot
plt.scatter(x,y,c='g')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('Graph in 2D')
plt.savefig('Scatter.png') ## saving the image
```

```python
## plot
##plt.plot(x,y,'r--')
## or 
plt.plot(x,y,'r',linestyle='dashed',linewidth=2,markersize=10)
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('Graph in 2D')
plt.savefig('plot.png')
```

```python
## subplots
plt.subplot(2,2,1) ##(row,col,pos)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
plt.savefig('subplots.png')
```

```python
## sin
x=np.arange(0,4*np.pi,0.1) ## (st,stop,step)
y=np.sin(x)
plt.title("sine wave Form")
plt.plot(x,y,'ro')
plt.savefig('sin.png')
```

```python
## sine and cosine
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,2,1)
plt.plot(x,y_sin)
plt.title("sine")
plt.subplot(2,2,2)
plt.plot(x,y_cos)
plt.title('cosine')
plt.savefig('sine&cosine.png')
plt.show()
```

```python
## Barplot
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.xlabel('Xaxis')
plt.ylabel('Yaxis')
plt.savefig('Barplot.png')
plt.show()
```

```python
## histogram
a=np.array([22,87,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("Histogram")
plt.savefig('Histogram.png')
```

```python
## Boxplot
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True)
plt.savefig('Boxplot.png')
```

```python
## pie chart
labels='python','c++','Ruby','Java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.4,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1F%%',shadow=True)
plt.axis('equal')
plt.savefig('pie.png')
```
